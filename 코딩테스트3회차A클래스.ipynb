# 문제 1. 문자열 나누기


image.png

```python
import time

start = time.time()

def solution(s):
# 답안 작성 부분 ===============
    answer = 0
    first_count = 0
    other_count = 0
    for i in s:
        if first_count == other_count:
            answer += 1
            k = i
        if i == k:
            first_count += 1
        else:
            other_count += 1
            



# ==============================
    return answer

end = time.time()

print(solution("banana"))
print(solution("abracadabra"))
print(solution("aaabbaccccabba"))

print("작동시간:", f"{end - start:.5f} sec")
```

# 문제 2. 둘만의 암호



image.png

```python
import time

start = time.time()

def solution(s, skip, index):
# 답안 작성 부분 ===============
    words = ''
    word = "abcdefghijklmnopqrstuvwxyz"
    for i in word:
        if i not in skip:
            words += i
    answer = ''
    for j in s:
        answer += words[(words.find(j)+index) % len(words)]     # find() 함수 : 문자의 인덱스 반환 / z -> a 다시 돌릴 때 [% len()] 사용하기
 



# ==============================
    return answer

end = time.time()

print(solution("aukks", "wbqd", 5))

print("작동시간:", f"{end - start:.5f} sec")
```

# 문제 3. 완주하지 못한 선수




image.png

```python
import time

start = time.time()

def solution(participant,	completion):
# 답안 작성 부분 ===============
    answer = ''
    participant.sort()
    completion.sort()
    for i in participant:
        if i in completion:
            participant.remove(i)
            completion.remove(i)
        else:
            answer += i
        





# ==============================
    return answer

end = time.time()

print(solution(["leo", "kiki", "eden"],	["eden", "kiki"]))
print(solution(["marina", "josipa", "nikola", "vinko", "filipa"],	["josipa", "filipa", "marina", "nikola"]))
print(solution(["mislav", "stanko", "mislavm", "ana"],	["stanko", "ana", "mislav"]))

print("작동시간:", f"{end - start:.5f} sec")
```

```python
import time

start = time.time()

def solution(X,Y):
# 답안 작성 부분 ===============
    answer = ''
    
    for i in range(9,-1,-1):
        answer += (str(i) * min(X.count(str(i)), Y.count(str(i))))
    
    if answer == '':
        return -1
    elif len(answer) == answer.count(str(0)):
        return 0
    else:
        return answer

        




# ==============================

end = time.time()

print(solution("100",	"2345"))
print(solution("100",	"203045"))
print(solution("100",	"123450"))
print(solution("12321",	"42531"))
print(solution("5525",	"1255"))

print("작동시간:", f"{end - start:.5f} sec")
```

# 문제 5. 햄버거 만들기


image.png
image.png

```python
import time

start = time.time()

def solution(ingredient):
# 답안 작성 부분 ===============
    answer = 0
    while True:
        found = False

        for i in range(len(ingredient)-3):
            if ingredient[i:i+4] == [1,2,3,1]:
                del ingredient[i:i+4]
                answer += 1
                found = True
                break
        if not found:
            break



# ==============================
    return answer

end = time.time()


print(solution([2, 1, 1, 2, 3, 1, 2, 3, 1]))
print(solution([1, 3, 2, 1, 2, 1, 3, 1, 2]))


print("작동시간:", f"{end - start:.5f} sec")
```
